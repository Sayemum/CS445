## Intro - Bike Rentals

Using a decision tree implementation, we will predict the number of bike rentals on the basis of time of day, time of year, and weather conditions.

## Part 1 - Hyper Parameter Tuning w/ Cross Validation

A hyperparameter for this decision tree is `max_depth` which is set by the programmer to see how the tree gets affected when changing it. Let's see what happens when we change the hyper parameter `max_depth`.

### Load Training & Testing Data (+accuracy score)

```python
import numpy as np
from decision_tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the training data
X_train = np.load('data/X_train.npy')
y_train = np.load('data/y_train.npy')

# Load the test data
X_test = np.load('data/X_test.npy')
y_test = np.load('data/y_test.npy')

max_depth = 3 # Change This Hyper-Parameter!

tree = DecisionTreeClassifier(max_depth=max_depth)
tree.fit(X_train, y_train)

y_pred = tree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Max Depth:", max_depth)
print("Accuracy:", accuracy)
```

According to the test above, we've created a decision tree classifier with a hyperparameter of `max_depth` that creates an accuracy score of the y testing data and the y predicted data. After doing some tuning with max_depth, it appears the more depth the tree has, the higher our accuracy is until it reaches a certain point where the accuracy doesn't improve by a lot. For example, the accuracy between `2` and `3` max_depth is `0.541` and `0.585` respectively. However, the accuracy between `10` and `20` max_depth is `0.773` and `0.735` respectively. So it's becoming less accurate after a certain max_depth has been reached.

## Part 2 - Model Evaluation (Test Set Classification Error)

```python
# Calculate classification error
classification_error = 1 - accuracy

print("Classification Error:", classification_error)
```

When identifying classification error, it's generally better to have a lower error, which signifies that the model is more accurate with its predictions. When testing a max_depth of 2 and 3, the classification error was 0.459 and 0.415 respectively. However, the classification error between 10 and 20 max_depth is 0.227 and 0.265 respectively. <b>This test indicates that small max_depth parameters are less accurate than higher params until they reach a certain point</b>. 2 and 3 have high errors. 10 and 20 have lower errors, but 20 ended up having a higher error.

## Part 3 - Classification Results (Test Data 4x4 matrix)

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate and print the confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

xlabels = ['0-190 rentals', '191-504 rentals', '505-1065 rentals', 'More than 1065 rentals']
ylabels = ['0-190 rentals', '191-504 rentals', '505-1065 rentals', 'More than 1065 rentals']

# Plot a heatmap of the matrix
plt.figure(figsize=(8, 6))

sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=xlabels, yticklabels=ylabels)

plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title("Classification Results")
plt.show()
```

Based on the confusion matrix illustrated as a heatmap above, each cell represents the number of samples with true labels predicted for each rental class. With a `max_depth` of `20`, some observations to note are that classes 0 and 3 have high true positives (346 and 357) which are good. On the other hand, classes 1 and 2 have less classification, with class 2 being the least classified. From this information, there may be an imbalance in the dataset or the model needs to find classes that are easier to predict. However, a `max_depth` of `3` surprisingly made classification harder for nearby classes. This may signify that reducing the depth leads to misclassification with the upside of being able to generalize data better.

## Part 4 - Attributes Used

We need to determine which attributes were used, so we'll create a function to calculate this for us.

```python
# Define the function to traverse the tree and collect attributes
def get_used_attributes(node, attributes):
    if node.split is not None:
        attributes.add(node.split.dim)
        if node.left is not None:
            get_used_attributes(node.left, attributes)
        if node.right is not None:
            get_used_attributes(node.right, attributes)

# Collect the attributes used in the tree
attributes_used = set()
get_used_attributes(tree._root, attributes_used)

# Print the attributes
attribute_names = ['Hour', 'Temperature(C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)',
                   'Dew point temperature(C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall(cm)',
                   'Seasons', 'Holiday', 'Functioning Day']

print("Attributes used:")
for attribute in attributes_used:
    print(attribute_names[attribute])
```

The max_depth hyperparameter can determine how many attributes are used in the tree. A max_depth of 20 uses all attributes, while 3 max_depth uses 5 attributes. Those attributes being `Hour, Temperature, Solar Radiation, Seasons, and Functioning Day`.
